```python
### Data input

pop_df = pd.read_csv('../../data/raw/ACSDT5Y2020.B08006_data_with_overlays_2022-04-29T134131.csv', skiprows=1) # modeshares (for commuting)
```

```python
#pop_df.loc[: , (pop_df != 'Margin of Error').all()]
filtered_cols = [col for col in pop_df.columns if 'Margin of Error' not in col]
filtered_cols = [col for col in filtered_cols if 'Male' not in col]
filtered_cols = [col for col in filtered_cols if 'Female' not in col]
filtered_cols
```

```python
pop_df = pop_df.filter(filtered_cols)
pop_df = pop_df.set_index('Geographic Area Name')
pop_df = pop_df.filter(like="Metro Area", axis=0)
pop_df.index = pop_df.index.str.replace(' Metro Area', '')
```

```python
### Create empty dataframe with ACS MSA names as index:
ms_df = pd.DataFrame(index = pop_df.index)
```

```python
ms_df.insert(loc = 0, column = 'ms_cara_pct', value = pop_df['Estimate!!Total:!!Car, truck, or van:!!Drove alone']/pop_df['Estimate!!Total:'])
ms_df.insert(loc = 1, column = 'ms_carp_pct', value = pop_df['Estimate!!Total:!!Car, truck, or van:!!Carpooled:']/pop_df['Estimate!!Total:'])
ms_df.insert(loc = 2, column = 'ms_pt_pct', value = pop_df['Estimate!!Total:!!Public transportation (excluding taxicab):']/pop_df['Estimate!!Total:'])
ms_df.insert(loc = 3, column = 'ms_taxi_pct', value = pop_df['Estimate!!Total:!!Taxicab, motorcycle, or other means']/pop_df['Estimate!!Total:'])
ms_df.insert(loc = 4, column = 'ms_bike_pct', value = pop_df['Estimate!!Total:!!Bicycle']/pop_df['Estimate!!Total:'])
ms_df.insert(loc = 5, column = 'ms_walk_pct', value = pop_df['Estimate!!Total:!!Walked']/pop_df['Estimate!!Total:'])
ms_df.insert(loc = 6, column = 'ms_wfh_pct', value = pop_df['Estimate!!Total:!!Worked from home']/pop_df['Estimate!!Total:'])
```
