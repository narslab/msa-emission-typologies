```python
cfc = pd.read_csv('../../data/tidy/msa-carbon-footprint.csv')
trf = pd.read_csv('../../data/tidy/tti-traffic.csv')
```

```python
trf_cfc = trf.merge(cfc, left_on='msa', right_on='msa', how='left')
```

```python
trf_cfc.head()
```

```python
g = sns.PairGrid(trf_cfc, dropna=True)
g.map_lower(sns.scatterplot)
g.map_diag(sns.histplot)
#plt.savefig('../../figures/pairwise-combined.png',dpi=300)
```
