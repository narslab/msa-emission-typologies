```python
### Data input

mods_df = pd.read_csv('../../data/raw/ACSDT5Y2019.B08006-Data.csv', skiprows=1) # modeshares (for commuting)
```

```python
#pop_df.loc[: , (pop_df != 'Margin of Error').all()]
filtered_cols = [col for col in mods_df.columns if 'Margin of Error' not in col]
filtered_cols = [col for col in filtered_cols if 'Male' not in col]
filtered_cols = [col for col in filtered_cols if 'Female' not in col]
filtered_cols = [col for col in filtered_cols if 'Annotation' not in col]
filtered_cols = [col for col in filtered_cols if 'Unnamed' not in col]
filtered_cols
```

```python
mods_df = mods_df.filter(filtered_cols)
mods_df = mods_df.set_index('Geographic Area Name')
mods_df = mods_df.filter(like="Metro Area", axis=0)
mods_df.index = mods_df.index.str.replace(' Metro Area', '')
```

```python
mods_df.index.name = 'msa'
```

```python
mods_df.columns = mods_df.columns.str.replace(':', '')
```

```python
### Create empty dataframe with ACS MSA names as index:
ms_df = pd.DataFrame(index = mods_df.index)
```

```python
ms_df.insert(loc = 0, column = 'ms_cara', value = mods_df['Estimate!!Total!!Car, truck, or van!!Drove alone']/mods_df['Estimate!!Total'])
ms_df.insert(loc = 1, column = 'ms_carp', value = mods_df['Estimate!!Total!!Car, truck, or van!!Carpooled']/mods_df['Estimate!!Total'])
ms_df.insert(loc = 2, column = 'ms_pt', value = mods_df['Estimate!!Total!!Public transportation (excluding taxicab)']/mods_df['Estimate!!Total'])
ms_df.insert(loc = 3, column = 'ms_taxi', value = mods_df['Estimate!!Total!!Taxicab, motorcycle, or other means']/mods_df['Estimate!!Total'])
ms_df.insert(loc = 4, column = 'ms_bike', value = mods_df['Estimate!!Total!!Bicycle']/mods_df['Estimate!!Total'])
ms_df.insert(loc = 5, column = 'ms_walk', value = mods_df['Estimate!!Total!!Walked']/mods_df['Estimate!!Total'])
ms_df.insert(loc = 6, column = 'ms_wfh', value = mods_df['Estimate!!Total!!Worked from home']/mods_df['Estimate!!Total'])
```
